```r
data <- read.csv("2022년_서울_일별평균대기오염도_2022.csv", fileEncoding="CP949")
head(data)
```

```r
san <- data %>%
    filter(grepl("산$", 측정소명)) %>%
    mutate(월 = substr(측정일시,5,6), 일 = substr(측정일시,7,8)) %>%
    mutate(월 = as.integer(월), 일 = as.integer(일)) %>%
    filter(월 == "11") %>%
    select(측정소명, 월, 일, 초미세먼지)

table(san$측정소명)
```

```r
any(is.na(san$초미세먼지))
```

```r
head(san)
```

```r
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p
```

```r
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p + labs(title = "2022년 11월 3지역 초미세먼지", x = "일", y = "초미세먼지")
```

```r
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p + labs(title = "2022년 11월 3지역 초미세먼지", x = "일", y = "초미세먼지")

p + theme(plot.title = element_text(color = "tomato", size = 30),
          axis.title.x = element_text(size = 20),
          axis.title.y = element_text(size = 20, color = 'blue'))
```

```r
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p + labs(title = "2022년 11월 3지역 초미세먼지", x = "일", y = "초미세먼지")

p + theme(plot.title = element_text(color = "tomato", size = 30),
          axis.title.x = element_text(size = 20),
          axis.title.y = element_text(size = 20, color = 'blue'))

p + ylim(c(0,80))
```

```r
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p + labs(title = "2022년 11월 3지역 초미세먼지", x = "일", y = "초미세먼지")

p + theme(plot.title = element_text(color = "tomato", size = 30),
          axis.title.x = element_text(size = 20),
          axis.title.y = element_text(size = 20, color = 'blue'))

p + scale_x_continuous(limits = c(0,31), breaks = 0:31) +
    scale_y_continuous(limits = c(0,80), breaks = seq(0,80,10))
```

```r
options(repr.plot.width = 7)
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p <- p + labs(title = "2022년 11월 3지역 초미세먼지",
         x = "일", y = "초미세먼지")
p + facet_wrap( ~ 측정소명, nrow = 3)
```

```r
options(repr.plot.width = 14)
p <- ggplot(san, aes(일, 초미세먼지, col = 측정소명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="2022년 11월 3지역 초미세먼지",
         x = "일", y = "초미세먼지")
p + facet_wrap( ~ 측정소명, ncol = 3)
```
