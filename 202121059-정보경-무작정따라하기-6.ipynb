```r
newdata <- data %>%
    mutate(부상자수 = 경상자수 + 중상자수 + 부상신고자수) %>%
    mutate(지역 = paste(시도, 시군구)) %>%
    filter(시도 == "경기")
          
grp <- group_by(newdata, 지역)           
result <- summarise(grp, 부상자수 = mean(부상자수))
head(newdata)
```

```r
summary(result$부상자수)
```

```r
bins <- (0:6) * 100
hist(result$부상자수, breaks = bins, col = 'gold')
```

```r
bins <- (0:6) * 100
tbl <- table(cut(result$부상자수, breaks = bins))
tbl
```

```r
hist(result$부상자수, breaks = bins, col = 'gold', xlim = c(-100, 700), axes = F)
mid <- seq(-50, 650, 100)
y <- c(0, tbl, 0)
lines(mid, y, type = 'o', col = 'red', lwd = 3)
box()
axis(side = 1, at = seq(-100,700,50), labels = seq(-100,700,50))
axis(side = 2, at = 0:10, labels = 0:10)
```
